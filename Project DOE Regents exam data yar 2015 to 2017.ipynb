```python
engl_df = sec_df.copy()
```

```python
#drop all the regents exams that are not english exams.
#first drop Regents Exams that are Nan
engl_df.dropna(subset = ["Regents Exam"], inplace=True)
#Use Regents Exams that are not English Regents
engl_df = engl_df[(engl_df['Regents Exam'] == 'Global History and Geography') |
        (engl_df['Regents Exam'] == 'Living Environment ') |
        (engl_df['Regents Exam'] == 'U.S. History and Government') |
        (engl_df['Regents Exam'] == 'English') |
        (engl_df['Regents Exam'] == 'Common Core English')]
engl_df.reset_index(inplace = True)
engl_df.head()
```

### To compare between years If each year increase or decrease on the total of students tested. Based on the data, the students are taking the regent exam less each year. 

```python
#Groupby with years
year_group = engl_df.groupby('Year')
#sum each  group
total_by_year_tested = year_group['Total Tested'].sum()
total_by_year_tested
```

To compare between years If each year increase or decrease on the total of students tested by year (only English subject). Upon looking at the bar chart I can immediately informed that less students are taking English regents each year. Less students are taking the regents each year. In fact, from 2015 to 2016 there were 91000 less students taking the English exam. There was an increased in 2015 but it decreased in 2017 by 130K.

```python
year = ['2015', '2016', '2017']
plt.figure(figsize=(5,5))
plt.bar(year, total_by_year_tested, color= 'steelblue')
plt.title ('Total Number of Students taking English Regents Exam  per year', fontsize = 14)
plt.xlabel('Year')
plt.ylabel('Total Number of Students')
```

Students are passing their regents (English) exam each year specially from year 2016 to 2017. I have observed that there was no data year 2017 for only English. 

```python
#Subject Passing Rate for Each Year
#Group by regents exam and year
group_year_regents = engl_df.groupby(by=['Regents Exam','Year'])
group_year_regents.size()
#Find the sum of total test
total = group_year_regents['Total Tested'].sum()
#Find the sum of total passed
total_pass = group_year_regents['Number Scoring 65 or Above'].sum()
#Find the passing rate of each group
dec_pass = total_pass.divide(total)
per_pass = dec_pass * 100
per_pass

fig, ax = plt.subplots(figsize=(15,10))
per_pass.unstack().plot.bar(ax=ax)
plt.title("Passing Rate of English Regents Per Year By Subject", fontsize = 14)
plt.ylabel('Percent of Students Passing')
_ = plt.xticks(rotation=0)
```

```python
#Subject Passing Rate for Each Year
#Group by regents exam and year
group_year_borough = engl_df.groupby(by=['Year','Borough'])
group_year_borough.size()
#Find the sum of total test
total = group_year_borough['Total Tested'].sum()
#Find the sum of total passed
total_pass = group_year_borough['Number Scoring 65 or Above'].sum()
#Find the passing rate of each group
dec_pass = total_pass.divide(total)
per_pass = dec_pass * 100
#Plot data in a bar graph
fig, ax = plt.subplots(figsize=(15,10))
per_pass.unstack().plot.bar(ax=ax)
plt.title("English Regents Passing Rate in Each Borough Per Year", fontsize = 14)
plt.ylabel('Percent of Students Passing')
_ = plt.xticks(rotation=0)
```
